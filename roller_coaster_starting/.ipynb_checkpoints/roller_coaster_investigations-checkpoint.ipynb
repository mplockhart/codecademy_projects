## Roller coaster
You will work to create several data visualizations that will give you insight into the world of roller coasters.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load rankings data here:
steel_df = pd.read_csv('Golden_Ticket_Award_Winners_Steel.csv')
wood_df = pd.read_csv('Golden_Ticket_Award_Winners_Wood.csv')
#print(steel_df.head())
wood_df.head()
```

```python
# look at the types of data
print(wood_df.dtypes)
print(steel_df.dtypes)
print(wood_df['Name'])
```

```python
# write function to plot rankings over time for 1 roller coaster here:
def rankings_plot(coaster_name, park_name, dataframe):
    name_park = dataframe[(dataframe['Park'] == park_name) & (dataframe['Name'] == coaster_name)]
    fig, ax = plt.subplots()
    plt.plot(name_park['Year of Rank'], name_park['Rank'])
    ax.invert_yaxis()
    ax.set_yticks([1,2,3])
    plt.title(f'Rank of {coaster_name} at {park_name} Theme Park')
    plt.xlabel('Year of Rank')
    plt.ylabel('Rank')
    plt.show()

rankings_plot('El Toro', 'Six Flags Great Adventure', wood_df)
    
plt.clf()
```

```python
# write function to plot rankings over time for 2 roller coasters here:
def rankings_plot(coaster_names_list, park_name_list, dataframe):
    name_park1 = dataframe[(dataframe['Park'] == park_name_list[0]) & (dataframe['Name'] == coaster_names_list[0])]
    name_park2 = dataframe[(dataframe['Park'] == park_name_list[1]) & (dataframe['Name'] == coaster_names_list[1])]
    fig, ax = plt.subplots()
    plt.plot(name_park1['Year of Rank'], name_park1['Rank'], label=coaster_names_list[0], color ='green')
    plt.plot(name_park2['Year of Rank'], name_park2['Rank'], label=coaster_names_list[1], color='red')
    ax.invert_yaxis()
    ax.set_yticks([1,2,3])
    plt.title('El Toro Vs Boulder Dash')
    plt.xlabel('Year of Rank')
    plt.ylabel('Rank')
    plt.legend()
    plt.show()

# Put the names you want to look up in the lists    
coaster_list = ['El Toro', 'Boulder Dash']
park_list = ['Six Flags Great Adventure', 'Lake Compounce'] 
rankings_plot(coaster_list, park_list, wood_df)

plt.clf()
```

```python
# write function to plot top n rankings over time here:

def top_n_ranked(selected_data_frame, rank):
    top_names = selected_data_frame['Name'].unique()
    top_list = []
    fig, ax = plt.subplots(figsize=(10,8))
    ticks = [num+1 for num in range(rank)]
    ax.invert_yaxis()
    ax.set_yticks(ticks)
    for coaster in top_names:
        top_list.append(selected_data_frame[selected_data_frame['Name'] == coaster])
    for coaster_num in range(len(top_list)):
        plt.plot(top_list[coaster_num]['Year of Rank'], top_list[coaster_num]['Rank'], label=top_list[coaster_num]['Name'].iloc[0])
    plt.legend(loc='lower right')
    plt.title(f'Top {rank} ranked roller coasters')
    plt.xlabel('Year of Rank')
    plt.ylabel('Rank')
    plt.show()


# for the top 5 ranked
max_rank = 5
top_ranked = wood_df[wood_df['Rank'] <= max_rank]
top_n_ranked(top_ranked, max_rank)

plt.clf()
```

```python
# load roller coaster data here:
rollers = pd.read_csv('roller_coasters.csv')
rollers = rollers.dropna().reset_index()
print(rollers.head())
#print(rollers.dtypes)
print(rollers.describe())

# write function to plot histogram of column values here:

def histogram_of_ints(dataframe):
    columns = dataframe.columns    
    for data in columns:
        if data == 'height':
            plt.hist(dataframe[data], range=[0, 100])
            plt.title(data)
            plt.show()
        else:
            plt.hist(dataframe[data])
            plt.title(data)
            plt.show()
    pass

ints_only = rollers.select_dtypes(include=['int', 'float'])
histogram_of_ints(ints_only)


plt.clf()
```

```python
# write function to plot pie chart of operating status here:
def operating_pie(dataframe):
    roller_status = dataframe[(dataframe['status'] == 'status.operating') | (dataframe['status'] == 'status.closed.definitely')]
    labels = ['Closed', 'Open']
    values = roller_status.groupby('status').size()
    plt.pie(values, labels=labels, autopct='%1.1f%%')
    plt.show()

operating_pie(rollers)



plt.clf()
```

```python
# write function to create scatter plot of any two numeric columns here:
def make_a_scatter(dataframe, xaxis, yaxis):
    xmax = np.percentile(dataframe[xaxis],[95])[0]
    xmax_df = dataframe[dataframe[xaxis] < xmax][[xaxis, yaxis]]
    ymax = np.percentile(xmax_df[yaxis],[95])[0]
    ymax_df = xmax_df[xmax_df[yaxis] < ymax]
    print(ymax_df)
    plt.scatter(ymax_df[xaxis], ymax_df[yaxis])
    plt.show()

make_a_scatter(rollers, 'height', 'speed')        
        
#print(rollers.dtypes)
#print(np.percentile(rollers['speed'], [75]))





plt.clf()
```
