```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels
import matplotlib.pyplot as plt
import math
```

```python
flight = pd.read_csv("flights.csv")
print(flight.head())
```

```python
## Task 1
sns.histplot(data=flight, x='coach_price')
plt.show()
plt.clf()
# From this data it seems that the mean data points are aroud $400. 
# Thius shows that $500 is towards the higher price range.
```

```python
## Task 2
hour8_flights = flight[flight['hours'] >= 8]
sns.histplot(data=hour8_flights, x='coach_price')
print(f"Mean flight cost is ${np.round(np.average(hour8_flights['coach_price']),2)}")
print(f"The minimum flight cost is ${np.round(np.amin(hour8_flights['coach_price']),2)}")
print(f"The maximum flight cost is ${np.round(np.amax(hour8_flights['coach_price']),2)}")
# $500 seems a more reasonable cost for an 8 hour ticket
plt.show()
plt.clf()
```

```python
## Task 3
sns.displot(data=flight, x='delay')
# Due to the massive difference looking at different x max values
plt.xlim(0,50)
print(f"The mean initial time of delays for the shorted mode is {np.average(flight[flight['delay'] <17]['delay'])}")
print(f"The mean initial time of delays for the longer mode is {np.average(flight[(flight['delay'] >17) & (flight['delay'] < 50)]['delay'])}")
plt.show()
plt.clf()
```

```python
## Task 4
flight_sub = flight.sample(frac=0.1)
sns.lmplot(x = "coach_price", y = "firstclass_price", data = flight_sub, line_kws={'color': 'black'}, lowess=True)
plt.show()
plt.clf()
```

```python
# Task 5
sns.histplot(data=flight, x='coach_price', hue='inflight_meal')
plt.show()
plt.clf()
sns.histplot(data=flight, x='coach_price', hue='inflight_entertainment')
plt.show()
plt.clf()
sns.histplot(data=flight, x='coach_price', hue='inflight_wifi')
plt.show()
plt.clf()
# now to look using box plots
sns.boxplot(data=flight, y='coach_price', x='inflight_meal')
plt.show()
plt.clf()
sns.boxplot(data=flight, y='coach_price', x='inflight_entertainment')
plt.show()
plt.clf()
sns.boxplot(data=flight, y='coach_price', x='inflight_wifi')
plt.show()
```

```python
## Task 6
sns.lmplot(x = "hours", y = "passengers", data = flight_sub, x_jitter = 0.25, scatter_kws={"s": 5, "alpha":0.2}, fit_reg = False)
plt.show()
plt.clf()
```

```python
sns.scatterplot(data=flight_sub, x='coach_price', y='firstclass_price', hue='weekend')
plt.show()
plt.clf()
```

```python
sns.boxplot(data=flight_sub, x='day_of_week', y='coach_price', hue='redeye')

plt.show()
plt.clf()
```
