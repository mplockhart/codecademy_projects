```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

# importing the data frames
matches_df = pd.read_csv('matches.csv')

# viewing the first few rows
#print(matches_df.head())

# printing information about the dfs
# matches_info = matches_df.info()
# matches_columns = matches_df.columns
# matches_shape = matches_df.shape

# working with mathces
# new data frame of home and away goals
matches_df['total_goals'] = matches_df['Home Team Goals'] + matches_df['Away Team Goals']
#print(matches_df['total_goals'])

# visualising goals each year (1930 - 2014) with a bar chart
f, ax = plt.subplots(figsize=(12,7))
sns.set_context("poster", font_scale = 1)
sns.set_theme(style="whitegrid")
ax = sns.barplot(data=matches_df, x='Year', y='total_goals')
ax.set_title('Goals scores from 1930 - 2014')
plt.show()
```

```python
# importing the goals data
goals_df = pd.read_csv('goals.csv')

# viewing the first few rows
# print(goals_df.head())

# printing information about the dfs
# goals_info = goals_df.info()
# goals_columns = goals_df.columns
# goals_shape = goals_df.shape

# make a goals boxplot
f, ax2 = plt.subplots(figsize=(12,7))
sns.set_context("poster", font_scale = 1.25)
sns.set_theme(style="whitegrid")
palette = sns.color_palette("Spectral")
sns.set_palette(palette)
ax2 = sns.boxplot(data=goals_df, x='year', y='goals')
ax2.set_title('Box plot showing sidtribution of goals per year')
plt.show()
```
